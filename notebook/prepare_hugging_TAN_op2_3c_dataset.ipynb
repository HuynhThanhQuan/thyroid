```python
import pandas as pd
import os
import shutil
```

```python
LOC = '../data/huggingface_TAN_opt2_3c'
```

```python
df = pd.read_csv('../metadata/TAN_opt2_3c.csv', index_col=0)
df
```

```python
df['image_id'] = df['image_path'].apply(lambda x : x.split('/')[-1])
```

```python
df['unique_image_id'] = df['patient_id'].astype(str) + '__' + df['new_label'].astype(str) + '__' +  df['image_id'] 
```

```python
os.makedirs(LOC, exist_ok=True)
```

```python
foldk = 'fold_4'
```

```python
gs = df.groupby([foldk,'new_label'])
```

```python
for i, g in gs:
    cur_folder = os.path.join(LOC, i[0], str(i[1]))
    os.makedirs(cur_folder, exist_ok=True)
    for _, r in g.iterrows():
        src_path = r['image_path']
        dst_path = os.path.join(cur_folder, r['unique_image_id'])
        shutil.copy(src_path, dst_path)
```
