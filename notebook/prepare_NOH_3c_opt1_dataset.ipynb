# PREPARE CSV FILE

```python
import pandas as pd
```

```python
df = pd.read_csv('../metadata/NOH_patient_dataset.csv',index_col=0)
df1 = pd.read_csv('../metadata/NOH_image_path.csv',index_col=0)
```

```python
df = df.merge(df1, how='inner', left_on='Patient # ', right_on='patient_id')
df
```

```python
df[['patient_id', 'label']].drop_duplicates()['label'].value_counts().sort_index()
```

```python
def new_label(x):
    if x == 0:
        return 0
    elif x <= 3:
        return 1
    else:
        return 2


df['new_label'] = df['label'].apply(lambda x: new_label(x))
df
```

```python
df[['patient_id', 'new_label']].drop_duplicates()['new_label'].value_counts().sort_index()
```

```python
df.to_csv('../metadata/NOH_opt1_3c.csv')
```
