# PREPARE CSV FILE

```python
import pandas as pd
```

```python
df = pd.read_csv('metadata/image_dataset_v2.csv',index_col=0)

def new_label(x):
    if x >= 3:
        return 2
    elif x >= 1:
        return 1
    else:
        return 0


df['new_label'] = df['label'].apply(lambda x: new_label(x))
df
```

```python
df['new_label'].value_counts(normalize=True)
```

```python
df.to_csv('metadata/image_dataset_3class.csv')
```

# COPY DATA

```python
import pandas as pd
import os
import shutil
```

```python
LOC = './data/huggingface_NOH_3class'
```

```python
df = pd.read_csv('./metadata/image_dataset_3class.csv', index_col=0)
df['unique_img_id'] = df['data'] + '__' + df['patient_id'] + '__' + df['label'].astype(str) + '__' + df['filename']
df
```

```python
df=df[df['data']=='NOH'].reset_index(drop=True)
df
```

```python
df['new_label'].value_counts(normalize=True)
```

```python
os.makedirs(LOC, exist_ok=True)
```

```python
gs = df.groupby(['dataset','new_label'])
```

```python
for i, g in gs:
    cur_folder = os.path.join(LOC, i[0], str(i[1]))
    os.makedirs(cur_folder, exist_ok=True)
    for _, r in g.iterrows():
        src_path = os.path.join('data', r['new_img_path'])
        dst_path = os.path.join(cur_folder, r['unique_img_id'])
        shutil.copy(src_path, dst_path)
```
